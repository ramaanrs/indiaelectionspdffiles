```python
import PyPDF2
import re
def extract_text_from_pdf(pdf_file:str) -> [str]:
    with open(pdf_file,'rb') as pdf:
        reader = PyPDF2.PdfReader(pdf,strict=False)
        pdf_text = []
      
        for page in reader.pages:
            content = page.extract_text_from_pdf()
            pdf_text.append(content)
            
        return pdf_text
    
if __name__ == '__main__':
    extracted_text = extract_text_from_pdf('F:/Ramaa/CoolResearch/IndiaElectionsStudies/LokSabhaElections/1951/1951(Vol-I).pdf')
    for text in extracted_text:
        split_message = re.split(r'\s+|[,;?!.-]\s*',text.lower())
        
        if 'constituency' in split_message:
            print(text)
```

```python
import PyPDF2

def convert_pdf_to_text(input_pdf_file,output_text_file):
    pdfFileObj = open(input_pdf_file,'rb')
    pdfReader = PyPDF2.PdfReader(pdfFileObj)
    pages = len(pdfReader.pages)
    #Loop to read all the pages
    for i in range(pages):
        pageObj = pdfReader.pages[i]
        pageno = ('Page' + str(i) + '_')
        text = pageObj.extract_text().split("\n")
        
        with open(output_text_file,'a') as f:
        # Iterate over the list and add page no and line number to each line
        for j in range(len(text)):
            line = pageno + 'Line' + str(j) + ':' + text[j] + "\n"
            f.write(line)
            
            return output_text_file           
```

```python
convert_pdf_to_text('1951(Vol-I).PDF','Text1951(Vol-I).txt')
```

```python
import PyPDF2

def convert_pdf_to_text(input_pdf_file,output_text_file):
    pdfFileObj = open(input_pdf_file,'rb')
    pdfReader = PyPDF2.PdfReader(pdfFileObj)
    pages = len(pdfReader.pages)
    #Loop to read all the pages
    for i in range(pages):
        pageObj = pdfReader.pages[i]
        text = pageObj.extract_text()
        
        with open(output_text_file,'a') as f:
            f.write(text)
            
            return output_text_file 
```
